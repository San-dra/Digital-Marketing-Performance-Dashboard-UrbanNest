```python
# --- GROUP SESSIONS BY MONTH_YEAR AND SAVE TO INDIVIDUAL GOOGLE SHEETS IN FOLDER ---
creds, _ = default()
gc = gspread.authorize(creds)
drive_service = build('drive', 'v3')

campaigns_sheet_name = "Campaigns_Master"
campaigns_folder_id = userdata.get('CAMPAIGNS_FOLDER_ID')
```

```python
# --- CREATE OR REPLACE Campaigns Sheet in Google Drive ---


# Step 1: Search for the file by name *inside the specific folder*
query = f"'{campaigns_folder_id}' in parents and name = '{campaigns_sheet_name}' and mimeType = 'application/vnd.google-apps.spreadsheet'"
results = drive_service.files().list(q=query, fields="files(id, name)").execute()
files = results.get('files', [])

if files:
    # If Found — open and overwrite
    file_id = files[0]['id']
    spreadsheet = gc.open_by_key(file_id)
    worksheet = spreadsheet.get_worksheet(0)
    worksheet.clear()
    print("Opened and cleared existing Campaigns_Master inside folder.")
else:
    # If Not found — create new and move
    spreadsheet = gc.create(campaigns_sheet_name)
    file_id = spreadsheet.id

    # Move to folder
    drive_service.files().update(
        fileId=file_id,
        addParents=campaigns_folder_id,
        removeParents='root',
        fields='id, parents'
    ).execute()
    worksheet = spreadsheet.get_worksheet(0)
    print("Created and moved new Campaigns_Master into folder.")

# Upload the DataFrame
gd.set_with_dataframe(worksheet, df_campaigns)
print("Campaign data written to Google Sheet.")

```
